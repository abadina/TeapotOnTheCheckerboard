# Calibration with planes

## Calibration with openCV

Importing modules

```python
import numpy as np
import cv2
import glob


# This is a value for the termination criterion of the subpixel corner localizer
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
```

Creation of the 3D point coordinates and loading images

```python
# Here you have 3D points for the checkerboard (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*7,3), np.float32)
objp[:,:2] = np.mgrid[0:7,0:6].T.reshape(-1,2)
# Here you load the image files
images = glob.glob('left*.jpg')

print(images)
```

Cell 1

We store the 3D points from the real word in array objpoints[] and 2D points in image place in array imgpoints[]for all the images we have and the loop over all the images , and in each loop we do those operations:
    -We use the function cvtColor() to convert the image from RGB to gray using this parameter cv2.COLOR_BGR2GRAY.
    -After converting the image to gray , we use the function findChessBoardCornes() to find the positions of internal corners of the chessboard . We pass in this function the gray image and the size of pattern . In other words,the number of inner corners per a chessboard row and column.And it returns the position of the corners  .
    - Depending on the return of the function below , if it is True (the input image is a view of the chessboard pattern ) , we add the objp defined in the beginning to the array objpoints[] and then we apply the function cornerSubPix() to refine the corner locations by using the gray image and the corners passed in parameters. Then we add the new corners to the array imgpoints[]


```python
# Arrays to store object points and image points from all the images.
objpoints = [] # 3d point in real world space
imgpoints = [] # 2d points in image plane.
for fname in images:
    #reading an image
    img = cv2.imread(fname)
    # transforming the image to gray scale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chess board corners
    ret, corners = cv2.findChessboardCorners(gray, (7,6),None)

    # If found, add object points, image points (after refining them)
    if ret == True:
        objpoints.append(objp)

        corners2 = cv2.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
        imgpoints.append(corners2)
        
print(gray)
```

Cell 2

By using the objpoints[] and imgpoints [] found in Cell 1 , we do the calibration using the function calibrateCamera() . We pass to this function objpoints[] , imgpoints [] and the gray image size. And it returns  ret, camera matrix, distortion coefficients, rotation and translation vectors.

```python
retval, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)

```

After reading the documentation of the cv2.calibrateCamera() we can say that we're using the calibration with planes method, that in addition to the fact that we're using the checkerboard images.

The output variables:
ret : the overall RMS re-projection error
mtx: camera matrix
dist: distortion coefficients


The output matrices:
rvecs: rotation  vectors
tvecs: translation vectors


```python
print("rotation vector size",len(rvecs))
```

rvecs represents the rotation vector and it’s an extrinsic parameter of the camera.
it is a convenient and most compact representation of a rotation matrix.
its size is (3x1) or (1,3) .

We can use the command Rodrigues() to retrieve R from rvecs.The function Rodrigues converts a rotation matrix to a rotation vector or vice versa. In our case ,we pass the vector of the rotation as a parameter (we can also pass an optional parameter Jacobian matrix) , it returns the matrix of rotation and a Jacobian matrix .


```python
Rm=[]
for rvec in rvecs:
    R,_=cv2.Rodrigues(rvec)
    Rm.append(R)
   # print(R)
   # print("\n")

```

Camera matrix refinement

```python
img = cv2.imread('left12.jpg')
h,  w = img.shape[:2]
newcameramtx, roi=cv2.getOptimalNewCameraMatrix(mtx,dist,(w,h),1,(w,h))
```

Cell 3

```python
# 1
dst = cv2.undistort(img, mtx, dist, None, newcameramtx)

# crop the image
x,y,w,h = roi
dst = dst[y:y+h, x:x+w]
cv2.imwrite('calibresult1.png',dst)


# 2
mapx,mapy = cv2.initUndistortRectifyMap(mtx,dist,None,newcameramtx,(w,h),5)
dst = cv2.remap(img,mapx,mapy,cv2.INTER_LINEAR)

# crop the image
x,y,w,h = roi
dst = dst[y:y+h, x:x+w]
cv2.imwrite('calibresult2.png',dst)
```

Radial distortion causes straight lines to appear curved. Radial distortion becomes larger the farther points are from the center of the image.
In cell 3 we worked on undistorting the image , to do so , we used two different methods.

The first method we used is the undistort method , and it’s the easiest to use . it returns the calibrated image.

The second method is using the remapping ,this way is a little bit more complicated. 
We should first, find a mapping function from the distorted image to the undistorted image. Then use the remap function. 

And as you can see, both the methods give the same result, all the edges are straight.


Cell 4

```python
tot_error = 0
for i in range(len(objpoints)):
    imgpoints2, _ = cv2.projectPoints(objpoints[i], rvecs[i], tvecs[i], mtx, dist)
    error = cv2.norm(imgpoints[i],imgpoints2, cv2.NORM_L2)/len(imgpoints2)
    tot_error += error

print("total error: ", tot_error/len(objpoints))
```

The quantity evaluated in cell 4 is the re-projection error.
It gives a good estimation of just how exact the found parameters are. The closer the re-projection error is to zero, the more accurate the parameters we found are.


## Teapot on the checkerboard

```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.animation
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.image as mpimg
```

Load teapot point cloud coordinates

```python
# Load from text file
tp= np.loadtxt("teapot.txt",usecols=range(3))
# Number of points in the cloud
n_points = np.shape(tp)
# Transpose and add a fourth coordinate with unitary value 
# (homogeneous coordinates)
tp = np.hstack((tp,np.ones((n_points[0],1),dtype=np.float))).T
# Save it in a data frame
df = pd.DataFrame({"x" : tp[0,:], "y" :tp[1,:], "z" : tp[2,:], 
                                                "w": tp[3,:]})
```

### Reminder

mtx : intrinsic matrix (Mint)

R: rotation matrix

t: translation matrix

[R t] : extrinsic matrix (Mext)


Camera matrix= Mint Mext

Calculate the extrinsic matrices

```python
Mext=[]
for i in range(len(Rm)):
    Rt_matrix = np.concatenate((Rm[i],tvecs[i]), axis=1)
    Mext.append(Rt_matrix)
    print(Rt_matrix)
    print("\n")
```

Building Camera matrices M.

```python
Ms=[]
for i in range(len(mtx)):
    M=np.matmul(mtx,Mext)
    Ms.append(M)
    print(M)
```

```python
# Scaling matrix
def scaling_matrix(s):
    # Scales are given in numpy array s
    # s=np.array([s_x,s_y,s_z])

    S = np.identity(4)
    S[:3, :3] = np.diag(s)
    
    return S
def translation_matrix(t):
    # Translation vector is given in numpy array t
    # t=np.array([t_x,t_y,t_z])
    
    # Write the translation matrix here
    T = np.identity(4)
    T[:3, 3] = t
    
    return T

# Rotation matrix
def rotation_matrix(theta, v):
    # Axis for rotation (unit norm vector) is given 
    # is given in numpy array v
    # v=np.array([v_x,v_y,v_z])
    # Rotation angle is given in theta
    
    # Write the rotation matrix here
    st = np.sin(theta)
    ct = np.cos(theta)
    o_ct = 1-ct
    vx2= v[0]**2
    vy2= v[1]**2
    vz2= v[2]**2
    vxy = v[0]*v[1]
    vxz = v[0]*v[2]
    vyz= v[1]*v[2]
    # Rotation
    R = [[ct+vx2*o_ct, vxy*o_ct-v[2]*st, vxz*o_ct+v[1]*st],
         [vxy*o_ct+v[2]*st, ct+vy2*o_ct, vyz*o_ct-v[0]*st],
         [vxz*o_ct-v[1]*st, vyz*o_ct+v[0]*st, ct+vz2*o_ct]]
    R_v = np.identity(4)
    R_v[:3, :3] = R
    
    return R_v
```

Teapot scaled

```python
objp = np.zeros((6*7,3), np.float32)
objp[:,:2] = np.mgrid[0:7,0:6].T.reshape(-1,2)
image = glob.glob('left1.jpg')

objpoints = [] # 3d point in real world space
imgpoints = [] # 2d points in image plane.

img = cv2.imread(image[0])
print(image[0])
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, corners = cv2.findChessboardCorners(gray, (7,6),None)
if ret == True:
    objpoints.append(objp)
    corners2 = cv2.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
    imgpoints.append(corners2)

retval, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)

R,_=cv2.Rodrigues(rvecs[0])

Mext = np.concatenate((R,tvecs[0]), axis=1)

camera_M=np.matmul(mtx,Mext)


```

### Animation

```python
def rotation_matrix2d(theta):
    sin = np.sin(theta)
    cos = np.cos(theta)

    # Rotation
    R = [[cos, -sin],
         [sin,  cos]]
    R_v = np.identity(3)
    R_v[:2, :2] = R
    
    return R_v

def scaling_matrix2d(s):

    S = np.identity(3)
    S[:2, :2] = np.diag(s)
    
    return S

def translation_matrix2d(t):

    T = np.identity(3)
    T[:2, 2] = t
    
    return T
```

```python
def update_graph2(num):
    # Load point cloud from data frame
    tp=np.vstack((df2['x'],df2['y'],df2['w']))
    angle = num*0.01
 
    M = scaling_matrix2d([1.0,1.0])

    #M = scaling_matrix([1.0,1.0,-1.0])
    # Compose M with rotation matrix
    M = np.dot(rotation_matrix2d(angle),M)
    
    # Compose M with translation matrix
    M = np.dot(translation_matrix2d(np.array([2.5,0.0])),M)
    M = np.dot(rotation_matrix2d(angle),M)
    # Apply composed transformation to the point cloud
    tp_r = np.dot(M,tp)
    # Update scatter plot
    graph._offsets2d = (tp_r[0,:], tp_r[1,:])
    return graph
```

```python
fig = plt.figure()
img=mpimg.imread('left1.jpg')
plt.imshow(img)
# 3D scatter plot
graph = plt.scatter(u_sim[0],u_sim[1],s=2)


# Animation function
ani = matplotlib.animation.FuncAnimation(fig, update_graph2,1000, 
                                            interval=10, blit=False)


plt.show()
```

### 3d Animation with transformation on teapot

```python
def update_graph3(num):
    # Load point cloud from data frame
    tp=np.vstack((df['x'],df['y'],df['z'],df['w']))
    angle = num*0.01
    M = scaling_matrix([0.5,0.5,-0.5])
    M = np.dot(rotation_matrix(angle, np.array([0.0,0.0,1.0])),M)
    M = np.dot(translation_matrix(np.array([2.5,0.0,0.0])),M)
    M = np.dot(rotation_matrix(angle, np.array([0.0,0.0,1.0])),M)
    # Apply composed transformation to the point cloud
    tp_r = np.dot(M,tp)
    # Update scatter plot
    graph._offsets3d = (tp_r[0,:], tp_r[1,:], tp_r[2,:])
    return graph

# 3D figure generation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
title = ax.set_title('Teapot')

# 3D scatter plot
graph = ax.scatter(tp[0,:], tp[1,:], tp[2,:],s=2)
# # Change axis limits if necessary
ax.set_xlim3d(-6, 6)
ax.set_ylim3d(-6, 6) 
ax.set_zlim3d(-1, 6)

# Animation function
ani = matplotlib.animation.FuncAnimation(fig, update_graph3,1000, 
                                            interval=1, blit=False)



plt.show()
```
